```python
sala1=["Sala 1",100,[],"Titanic"]
sala2=["Sala 2",60,[],"Fast and Furious"]
sala3=["Sala 3",50,[],"Oppenheimer"]
sala4=["Sala 4",40,[],"Creed"]

cinema = [sala1,sala2,sala3,sala4]

def listar(cinema):
    print("***Catálogo***")
    for sala in cinema:
        print(sala[0] + ":" + sala[3])
    return 

def disponivel(cinema,filme,lugar):
    disponivel = False
    for sala in cinema:
        if filme == sala[3] and lugar not in sala[2]:
            disponivel = True
    return disponivel 

def vendeBilhetes(cinema,filme,lugar):
    for sala in cinema:
        if filme == sala[3]:
            if disponivel(cinema,filme,lugar):
                sala[2].append(lugar)
                print(f"""O lugar {lugar} foi reservado - Bilhete: {sala}""")
            else:
                print("Este lugar já se encontra ocupado")
    return

def listardisponibilidades(cinema):
    print("Salas ainda disponíveis:")
    for sala in cinema:
        n = sala[1] - len(sala[2])
        print(f"{sala[0]} com {n} lugares disponíveis")
    return

def inserirSala(cinema,sala,lugares,filme):
    s = [sala,lugares,[],filme]
    if s not in cinema:
        cinema.append(s)
        print(f"Sala adicionada com sucesso: {s} ")

    else:
        print("Esta sala já existe")

def menu():
    print("""Bem-vindo ao nosso cinema.
Siga os próximos passos para escolher o seu filme:
(1) Filmes em exibição;
(2) Disponibilidade das sala e dos lugares;
(3) Comprar bilhete, somente após verificar a disponibilidade;
(4) Listar os lugares livres da sala;
(5) Adicionar uma sala ao cinema;    
(6) Fechar o menu""")
    
menu()

while menu != 6:
    print("--------------------------------------")
    menu = int(input("Escolha uma opção:"))
    if menu == 1:
        listar(cinema)
    elif menu == 2:
        print(disponivel(cinema,input("Escolhe o filme que queres ver"),int(input("Escolhe um lugar da sala"))))
    elif menu == 3:
        vendeBilhetes(cinema,input("Escolhe o filme que queres ver"),int(input("Escolha o lugar que quer: ")))
    elif menu == 4:
        listardisponibilidades(cinema)
    elif menu == 5:
        inserirSala(cinema,input("Qual o nome da nova sala?"),int(input("Qual a lotação da sala?")),input(("Qual o filme em exibição?")))

print("Obrigado por escolher o nosso cinema!!!")











```
