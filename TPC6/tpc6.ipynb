```python
# TabMeteo = [(Data,TempMin,TempMax,Precipitacao)]
    # Data = (Int,Int,Int)
    # TempMin = Float
    # TempMax = Float
    # Precipitacao = Float

tabMeteo = [((2022,1,20), 2, 16, 0), ((2022,1,21), 1, 13, 0.2), ((2022,1,23), 6, 19, 0.6), ((2022,1,24), 3, 18, 0.8), ((2022,2,20), 6, 19, 0.2), ((2022,2,24), 3, 18, 0.2), ((2022,2,28), 3, 18, 0.2)]

def menu():
    print("""---------------MENU---------------
1: Temperatura média diária
2: Temperatura mais baixa registada
3: Amplitude térmica diária
4: Dia mais chuvoso
5: Sequencia de dias chuvosos acima de p
6: Sequência de dias com menos chuva (abaixo de p)
7: Guardar tabela meteorológica num ficheiro
8: Carregar tabela meteorológica de um ficheiro
9: Gráficos da temperatura mínima, máxima e pluviosidade
0: Sair
----------------------------------------------------------""")

def medias(tabMeteo):
    res = []
    for data,tmin,tmax,prec in tabMeteo:
        media = (tmin + tmax) / 2
        res.append((data,media))
    return res

def minMin(tabMeteo):
    min = tabMeteo [0][1]
    for i in tabMeteo:
        if i[1] < min:
            min = i[1]
    return min

def amplTerm(tabMeteo):
    res = []
    for t in tabMeteo:
        data,tmin,tmax,perc = t
        soma = tmax - tmin
        res.append((data,soma))        
    return res  

def maxChuva(tabMeteo):
    max_prec = tabMeteo[0][3]
    for i in tabMeteo:
        if i[3] > max_prec:
            max_prec = i[3]
            max_data = i[0]
    return (max_data, max_prec)

def diasChuvosos(tabMeteo, p):
    res = []
    for t in tabMeteo:
        if t[3] > p:
            n = (t[0], t[3])
            res.append(n)
    return res

def maxPeriodoCalor(tabMeteo, p):
    cont = 0
    for t in tabMeteo:
        if t[3] < p:
            cont = cont + 1
        else:
            cont = 0   
    return cont

def guardaTabMeteo(t, fnome):
    f = open(fnome, "w")
    res = ""
    for data,tmin,tmax,prec in t:
        ano, mes, dia = data
        f.write(f"{ano}-{mes}-{dia};{tmin};{tmax};{prec}\n")
    f.close()
    return

def carregaTabMeteo(fnome):
    res = []
    file = open(fnome,"r")
    for line in file:
        #line = line[:-1]
        line = line.strip()
        campos = line.split(";")
        data,tmin,tmax,prec = campos
        ano, mes, dia = data.split("-")
        tuplo = ((int(ano), int(mes), int(dia)),float(tmin),float(tmax),float(prec))
        res.append(tuplo)
    file.close()
    return res

from matplotlib import pyplot as plt

def grafTabMeteo(t):
    x = [f"{data[0]}-{data[1]}-{data[2]}" for data, tmin, tmax, prec in t]
    ytmin = [tmin for data, tmin, tmax, prec in t]
    ytmax = [tmax for data, tmin, tmax, prec in t]
    
    yprec = [prec for *_, prec in t]   # *_ -> serve para simplificar a escrita dos que não são usados

    plt.plot(x,ytmin, label = "Temperatura Mínima (ºC)", color="blue", marker= "o")
    plt.plot(x,ytmax, label = "Temperatura Máxima (ºC)", color="red", marker= "o")
    plt.legend()
    plt.title("Tabela Meteorológica")
    plt.grid()
    plt.xticks(rotation=45)
    plt.show()

    plt.bar(x, yprec, label = "Pluviosidade (mm)", color = "c")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid
    plt.show()
    return

modo = -1

menu()

while modo != 0:
    modo = int(input("Escolha uma das opções: "))
    if modo == 1:
        print("Temperatura média diária:", medias(tabMeteo))
    elif modo == 2:
        print(f"Temperatura mais baixa registada: {minMin(tabMeteo)}")
    elif modo == 3:
        print(f"Amplitude térmica diária: {amplTerm(tabMeteo)}" )
    elif modo == 4:
        print(f"Dia mais chuvoso: {maxChuva(tabMeteo)}")
    elif modo == 5:
        p = float(input("Introduza o valor mínimo de precipitação:"))
        print(f"Sequencia de dias chuvosos acima de {p}: {diasChuvosos(tabMeteo, p)}")
    elif modo == 6:
        p = float(input("Intoduza o valor máximo de precipitação: "))
        print(f"Sequência de dias com menos chuva (abaixo de {p}): {maxPeriodoCalor(tabMeteo, p)}")
    elif modo == 7:
        fnome = input("Introduza um nome para o ficheiro:")
        guardaTabMeteo(tabMeteo, fnome)
        print("Tabela meteorológica guardada num ficheiro com sucesso")
    elif modo == 8:
        fnome = input("Nome do ficheiro:")
        print(carregaTabMeteo(fnome))
    elif modo == 9:
        print("GRÁFICOS")
        grafTabMeteo(tabMeteo)
    elif modo == 0:
        print("Aplicação encerrada")
```
