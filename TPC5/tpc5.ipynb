```python
def menu():
    print("""----------------- MENU -----------------
1: Criar uma turma
2: Inserir um aluno na turma
3: Listar a turma
4: Consultar um aluno por ID
5: Guardar a turma em ficheiro
6: Carregar a turma dum ficheiro
0: Sair
----------------------------------------""")

def criarTurma():
    turma = []
    print("A sua turma foi criada")
    return turma

def inserirAluno(listaTurma):
    nome = input("Introduza o nome do aluno: ")
    id = int(input("Introduza o id do aluno"))
    nota_1 = int(input("Introduza a nota do aluno nos TPCs: "))
    nota_2 = int(input("Introduza a nota do aluno no projeto: "))
    nota_3 = int(input("Introduza a nota do aluno no teste : "))
    notas = [nota_1,nota_2,nota_3]
    aluno = (nome, id, notas)
    listaTurma.append(aluno)
    print(listaTurma)
    return listaTurma

def listarTurma(listaTurma):
    print(f"Lista de alunos na turma: {listaTurma}")

def idAluno(listaTurma, i):
    for t in listaTurma:
        if i == t[1]:
            print(f"O aluno com id {i} é o/a {t}")
        else:
            print("Aluno não encontrado.")
    return

def guardarTurma(listaTurma, fnome):
    file = open(fnome, "w")
    res = ""
    for a in listaTurma:
        aluno, id, notas = a
        res = res + str(aluno) + ";" + str(id) + ";" + str(notas[0]) + "::" + str(notas[1]) + "::" + str(notas[2]) + "|"
        res = res + "\n"
    file.write(res)
    file.close()

def carregarTurma(fnome):
    listaTurma = [] 
    file = open(fnome, "r")
    text = file.read()
    turma_text = text.split("\n")

    for p_text in turma_text[:-1]:
        t = []
        t_text = p_text.split("|")

        for t_text in t_text[:-1]:
            aluno, id, notas = t_text.split(";")

            notas_lista = []
            for n in notas.split("::"):
                notas_lista.append(float(n))
            al = (str(aluno),int(id),list(notas_lista))
            t.append(al)
        listaTurma.append(t)
    file.close()
    return listaTurma

modo = -1

menu()

while modo != 0:
    modo = int(input("Escolha a opção que deseja efetuar: "))
    if modo == 1:
        turma = criarTurma()
    elif modo == 2:
        inserirAluno(turma)
    elif modo == 3:
        listarTurma(turma)
    elif modo == 4:
        identidade = int(input("Introduza o id do aluno que procura: "))
        idAluno(turma,identidade)
    elif modo == 5:
        print("A sua turma foi guardada com sucesso na ficheiro: turma")
        guardarTurma(turma, "turma.txt")
    elif modo == 6:
        print(carregarTurma("turma.txt"))
    elif modo == 0:
        print("Aplicação encerrada")

```
